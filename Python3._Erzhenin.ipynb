Полученное число - есть номер страницы, по материалам которой он задаст вопрос студенту. Помогите преподавателю реализовать такой алгоритм.

Примечание:
*   Порядок номеров в исходном списке может быть любым.
*   Поиск максимального значения из списка осуществить с помощью цикла while



### Пример

**Ввод:**

```
[33, 15, 17, 13, 92, 54, 29, 8]
```

**Вывод:**

```
11
```

````python
list_of_elements = [33, 15, 17, 13, 92, 54, 29, 8]
````

````python
# ваш код в этой ячейке
pages = [int(num) for num in input("Введите номера страниц через пробел: ").split()]

# Находим максимальное значение в списке с помощью цикла while
max_page = pages[0]  
i = 1
while i < len(pages):
    if pages[i] > max_page:
        max_page = pages[i]
    i += 1

# Вычисляем номер страницы, по которой будет вопрос
question_page = max_page // len(pages)

# Вывод результата
print("Номер страницы для вопроса:", question_page)
````

## Задача 5. Всё по порядку

Дано натуральное число. Напишите программу, которая определяет, является ли последовательность его цифр при просмотре слева направо упорядоченной по убыванию. Если условие удовлетворено выведите `ДА`, в противном случае - `НЕТ`

### Пример

**Ввод:**

```
75320
```

**Вывод:**

```
ДА
```

````python
# дополните код в этой ячейке
num_for_check =input("Введите натуральное число: ")
````

````python
# ваш код в этой ячейке
num_for_check =input("Введите натуральное число: ")

is_decreasing = True
for i in range(len(num_for_check) - 1):
    if int(num_for_check[i]) < int(num_for_check[i + 1]):
        is_decreasing = False
        break

# Выводим результат
print ("Введенное число:", num_for_check)
if is_decreasing:
    print("ДА")
else:
    print("НЕТ")
````

## Задача 7. Затроилось...

Попросите пользователя ввести какое-либо слово, а также число. При помощи циклов выведите каждый символ строки, при этом символ должен повторяться количество раз равным числу, что ввел пользователь. Каждый последующий новый символ необходимо выводить с новой строки, например: ок.

### Пример

**Ввод:**

```
Привет
3
```

**Вывод:**

```
ППП
ррр
иии
ввв
еее
ттт
```

````python
# ваш код в этой ячейке
word = input("Введите слово: ")
number = int(input("Введите число: "))

print("Введенное слово:", word)
for char in word:
    for _ in range(number):
        print(char, end="")
    print() 
````

## Задача 8. Таракан

Студент в общежитии проводит перепись своих вещей. Он до смерти боится тараканов, поэтому, только увидев их, бросает все свои дела и убегает. Необходимо вывести на экран вещи из списка студента, но если среди вещей появится '`таракан`' — немедленно выйти из цикла и на экран вывести '`Студен убежал'`

Примечение: необходимо предлагать ввод вещей с клавиатуры до момента ввода слова `таракан`.

### Пример

**Ввод:**

```
учебник
тетрадь
грязная кружка
степлер
носки
таракан
```

**Вывод:**

```
учебник
тетрадь
грязная кружка
степлер
носки
Студент убежал
```

````python
# ваш код в этой ячейке
things = []
while True:
    thing = input("Введите вещь студента: ")
    if thing.lower() == "таракан":
        break
    things.append(thing)


for thing in things:
    print(thing)

print("Студент убежал")
````
